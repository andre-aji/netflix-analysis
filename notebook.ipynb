```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

Subsetting the movies released in the 1990s

```python
netflix_movies = netflix_df[netflix_df["type"] == "Movie"]
movies_1990 = netflix_movies[(netflix_movies["release_year"] >= 1990) & (netflix_movies["release_year"] < 2000)]
```

Finding the Distribution of Durations

```python
plt.hist(movies_1990["duration"])
plt.title("1990s Movies")
plt.xlabel("Duration in Minutes")
plt.ylabel("Amount of Movies")
plt.show()
```

subsetting only action movies from the 1990s

```python
actionmovies_1990 = movies_1990[movies_1990["genre"] == "Action"]
```

Creating a counter for short action movies

```python
short_action_movies = 0
```
